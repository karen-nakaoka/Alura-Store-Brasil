# 📊 **Relatório Final – Análise das Lojas da Alura Store**

O objetivo desta análise foi avaliar o desempenho das quatro lojas da rede Alura Store, a fim de recomendar qual delas deve ser vendida para viabilizar um novo investimento do Senhor João.

Para isso, foram analisados os seguintes indicadores:

* Faturamento total

* Ticket médio

* Categorias mais e menos vendidas

* Produtos mais e menos vendidos

* Média de avaliação dos clientes

* Frete médio

A partir dessas métricas e dos gráficos gerados, foi possível comparar o desempenho das lojas de forma quantitativa e estratégica.


📈 **Desenvolvimento da Análise**


**Faturamento Total**

A Loja 1 apresentou o maior faturamento (`R$ 1.534.509`), seguida pelas Lojas 2 e 3. A Loja 4 apresentou o menor faturamento (`R$ 1.384.497`).

Esse resultado indica que, sob o ponto de vista puramente financeiro, a Loja 4 possui o menor desempenho em receita total.


**Ticket Médio**

A Loja 4 também apresentou o menor ticket médio (`R$ 587,15`), aproximadamente 9,7% abaixo da Loja 1, que teve o maior (`R$ 650,49`).

Isso ajuda a explicar o menor faturamento da Loja 4, indicando que seus produtos possuem, em média, menor valor unitário.


**Categorias e Produtos**

Em todas as lojas, a categoria mais vendida foi Móveis, demonstrando padrão semelhante de mix de produtos.

A distribuição dos produtos mais e menos vendidos mostrou volumes relativamente equilibrados entre as lojas, sem evidência de estoque problemático ou baixa saída significativa em uma loja específica.

Portanto, não há indícios de que o desempenho inferior da Loja 4 esteja relacionado a falhas no mix de produtos.


**Avaliação Média dos Clientes**

A Loja 3 apresentou a melhor média de avaliação (4,05), seguida da Loja 2 (4,04).

A Loja 4 apresentou média de 3,99, enquanto a Loja 1 obteve a menor média (3,98).

Apesar de ter menor faturamento, a Loja 4 não apresenta pior satisfação dos clientes, o que indica que sua performance não está relacionada a problemas de qualidade ou experiência.


**Frete Médio**

A Loja 1 apresentou o maior frete médio (`R$ 34,69`), enquanto a Loja 4 apresentou o menor (`R$ 31,28`).

Esse dado sugere que a Loja 4 possui maior eficiência logística ou menor custo médio de entrega, o que pode ser um ponto positivo competitivo.


🧠 **Conclusão e Recomendação**

Considerando todos os fatores analisados, a Loja 4 apresenta:

* O menor faturamento total

* O menor ticket médio

* Boa avaliação dos clientes

* Menor frete médio

* Mix de produtos semelhante às demais lojas

Embora a Loja 4 apresente pontos positivos, como boa avaliação e menor custo de frete, seu desempenho financeiro é inferior às demais, principalmente em relação ao faturamento total e ticket médio.

Dessa forma, sob uma perspectiva financeira e de maximização de receita para reinvestimento, _recomenda-se a venda da Loja 4_, por apresentar o menor retorno econômico dentro do conjunto analisado.

No entanto, é importante destacar que a Loja 4 demonstra potencial estratégico, especialmente em eficiência logística e satisfação do cliente, o que poderia ser explorado caso o objetivo fosse expansão e melhoria operacional, e não apenas geração imediata de capital.
